# Data Cleaning:
- in this section we will deal with outliers,skewness, weak correlation etc.

```python
#removing categorical feature variables

df.drop(categorical,axis=1,inplace=True)
```

```python
print(df['EmployeeCount'].value_counts())
print(df['StandardHours'].value_counts())
print(df['Over18_New'].value_counts())
```

```python
#so these feature variables have only one type of class which is same for all the employees, so it won't be very useful for our model
df.drop(['EmployeeCount','StandardHours','Over18_New'],axis=1,inplace=True)
```

```python
df.shape
```

```python
#let's try zscore method for outliers removal

from scipy.stats import zscore

z=np.abs(zscore(df))
np.where(z>3)
```

```python
z.iloc[28][23]
```

```python
#dropping the weakly correlated variables

df_new2=df_new.drop(['WorkLifeBalance','TrainingTimesLastYear','DailyRate','RelationshipSatisfaction','YearsSinceLastPromotion','Education','YearsSinceLastPromotion','EmployeeNumber','PercentSalaryHike','HourlyRate','BusinessTravel_New','PerformanceRating','MonthlyRate','EducationField_New','Gender_New','NumCompaniesWorked','Department_New','JobRole_New','DistanceFromHome'],axis=1)
df_new2.shape
```

```python
# now selecting our feature and target variable for further process

x=df_new2.drop('Attrition_New',axis=1)
y=df_new2['Attrition_New']
```

```python
#data transformation for less skewness

from sklearn.preprocessing import power_transform

x=power_transform(x,method='yeo-johnson')
```

```python
#checking for multicollinearity

from statsmodels.stats.outliers_influence import variance_inflation_factor

def vif_calc():
    vif=pd.DataFrame()
    vif['VIF Factor']=[variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
    vif['features']=x.columns
    print(vif)
```

```python
vif_calc()
```

- okay so only three features have vif above 5 that too not so much, multicollinearity isn't very strong between features

```python
#now

y.value_counts()
```

- it can be seen that the data is highly imbalanced, so we will use over sampling method

```python
from imblearn.over_sampling import SMOTE

smt=SMOTE()

train_x,train_y=smt.fit_resample(x,y)
```

```python
train_x.shape
```

```python
train_y.shape
```

```python
train_y.value_counts()
```

- okay so data is balanced now

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
lr=LogisticRegression()
```

```python
for i in range(0,100):
    x_train,x_test,y_train,y_test=train_test_split(train_x,train_y,random_state=i,test_size=.30)
    lr.fit(x_train,y_train)
    predtrain=lr.predict(x_train)
    predlr=lr.predict(x_test)
    if round(accuracy_score(y_test,predlr)*100)==round(accuracy_score(y_train,predtrain)*100):
        print('At random state',i,'the model performs very well')
        print('At random state',i,':')
        print('Training accuracy score is',accuracy_score(y_train,predtrain)*100)
        print('Testing accuracy score is',accuracy_score(y_test,predlr)*100,'\n')
```

```python
#so we will use random state as 19 for further process

x_train,x_test,y_train,y_test=train_test_split(train_x,train_y,random_state=19,test_size=.30)
lr.fit(x_train,y_train)
predtrain=lr.predict(x_train)
predlr=lr.predict(x_test)
```

```python
print(accuracy_score(y_test,predlr)*100)
print(confusion_matrix(y_test,predlr))
print(classification_report(y_test,predlr))
```

```python
from sklearn.metrics import mean_squared_error

mselr=mean_squared_error(y_test,predlr)

print(mselr)
```

```python
#checking cross validation score for logistic regression

for j in range(2,10):
    lrs=cross_val_score(lr,train_x,train_y,cv=j)
    lrc=lrs.mean()
    print('At cv=',j)
    print('Cross Validation Score is:',lrc*100)
    print('Accuracy score is:',accuracy_score(y_test,predlr)*100,'\n')
```

- as cross validation score coming out to be very close to accuracy score the model is working fine

```python
#roc auc plotting for logisticRegression

from sklearn.metrics import roc_curve,auc

fpr,tpr,thresholds=roc_curve(y_test,predlr)

roc_auc=auc(fpr,tpr)

plt.figure()
plt.plot(fpr,tpr,color='darkorange',lw=10,label='ROC curve (area=%.2f)'%roc_auc)
plt.plot([0,1],[0,1],color='navy',lw=10,linestyle='--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel('False Positive rate')
plt.ylabel('True Positive rate')
plt.title('Receiver Operating Characteristics')
plt.legend(loc='lower right')
plt.show()
```

- so our model is working well than the base model
